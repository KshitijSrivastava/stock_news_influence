```python
#Create pickle files for data containing stock price

import csv
from datetime import datetime
import pickle


def save_charts_in_pickle(file_dir):

    stock_details = {}
    with open(file_dir) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            dt = datetime.strptime( row[0] + ":" + row[1] , "%Y.%m.%d:%H:%M")
            start_stock = row[2]
            end_stock = row[3]
            if float(end_stock) - float(start_stock) >= 0:
                direction = 1
            else:
                direction = -1
            stock_details[dt.date()] = [start_stock, end_stock, direction ]
        
        pickle_file = "pickle_charts/APPLE1440"
        dbfile = open(pickle_file, 'wb')
        pickle.dump(stock_details, dbfile)
        dbfile.close()




save_charts_in_pickle("CHARTS/APPLE1440.csv")

save_charts_in_pickle('CHARTS/AMAZON1440.csv')
```

```python
#For creating a folder with a name

import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
#Arranging the news articles by the news site

import json
from datetime import datetime

data = {}

for i in range(1, 1414):
    try:
        file_number = "{:07d}".format(i)
        f = open("News/2019_02_d157b48c57be246ec7dd80e7af4388a2/news_{}.json".format(file_number), "r")
        json_object = json.loads(f.read())
        url = json_object["url"]
        text = json_object["text"]
        site = json_object["thread"]["site"]
        published = json_object.get("published", "")
        
        news_obj = {}
        news_obj["url"] = url
        news_obj["text"] = text
        news_obj["site"] =  site
        news_obj["published"] = published
        
        if site not in data:
            data[site] = [news_obj]
        else:
            data[site].append(news_obj)
            
    except:
        print("Failure !!! at {}".format(i))
```

```python
#Storing the news by the folder name and month

foldername = "2019_02_d157b48c57be246ec7dd80e7af4388a2" + "_ordersites"
createFolder('./'+ foldername +'/')
for site in data.keys():
    with open( "{}/{}.json".format(foldername, site), "w") as outfile:
        json_object = json.dumps(data[site])

        outfile.write(json_object)
```

```python
#Loads the pickle files for stock chart data for Amazon and Apple

import pickle

amazon_pickle_file = open ("pickle_charts/AMAZON1440", "rb")
amazon_stock_charts = pickle.load(amazon_pickle_file)

apple_pickle_file = open ("pickle_charts/APPLE1440", "rb")
apple_stock_charts = pickle.load(apple_pickle_file)

```

```python
import pysentiment2 as ps

def find_text_polarity(txt_list):
    """
    Finds the polarity of the text using McDonald Financial Sentiment Dictionaries
    """
    lm = ps.LM()
    tokens = lm.tokenize(". ".join(txt_list))
    score = lm.get_score(tokens)
    return score["Polarity"]

    
```

```python
from datetime import datetime
import pysentiment2 as ps


amazon_keywords = [ "amazon.com", "amazon", "amzn", "aws", "amazon web service", "$amzn"]
apple_keywords = [ "apple.com" "apple", "appl", "iphone", "ipad", "mac", "macbook", "$appl"]

def find_text_prediction_amazon(amazon_lines, date_time):
    """
    amazon_lines: A list containing strings relevant to Amazon
    date_time: Datetime object
    """
    polarity = find_text_polarity(amazon_lines)
    try:
        stock_details = amazon_stock_charts[date_time.date()]
    except:
        return 0
    
    if (polarity > 0 and stock_details[2] == 1) or (polarity < 0 and stock_details[2] == -1):
        return 1
    elif (polarity > 0 and stock_details[2] == -1) or (polarity < 0 and stock_details[2] == 1):
        return -1
    else:
        return 0
    
    
def find_text_prediction_apple(apple_lines, date_time):
    """
    apple_lines: A list containing strings relevant to Apple
    date_time: Datetime object
    """
    polarity = find_text_polarity(apple_lines)
    try:
        stock_details = apple_stock_charts[date_time.date()]
    except:
        return 0
    
    if (polarity > 0 and stock_details[2] == 1) or (polarity < 0 and stock_details[2] == -1):
        return 1
    elif (polarity > 0 and stock_details[2] == -1) or (polarity < 0 and stock_details[2] == 1):
        return -1
    else:
        return 0
        
        
    
def news_article_accuracy(obj):
    """
    To find the prediction accuracy of the article
    obj: contains a dict of keys -> text, published, site etc
    """
    text = obj["text"]
    published_datetime = obj['published']
    date_time = datetime.strptime(published_datetime[:-6], "%Y-%m-%dT%H:%M:%S.%f")
    
    amazon_lines = []
    apple_lines = []
    for l in text.split(". "):
        line = l.lower()
        for amazon_token in amazon_keywords:
            if amazon_token in line:
                amazon_lines.append(line)
        for apple_token in apple_keywords:
            if apple_token in line:
                apple_lines.append(line)
                 
    amazon_lines = list(set(amazon_lines))
    apple_lines = list(set(apple_lines))
    
    if len(amazon_lines) != 0:
        amazon_accuracy = find_text_prediction_amazon(amazon_lines, date_time)
    else:
        amazon_accuracy = 0
        
    if len(apple_lines) != 0:
        apple_accuracy = find_text_prediction_apple(apple_lines, date_time)
    else:
        apple_accuracy = 0
                 
    return (amazon_accuracy, apple_accuracy)
    
    
```

```python
import os
import json

folder_name = [  "2018_" + "{:02d}".format(i) + "_ordersites" for i in range(1, 7)  ]
#folder_name = []
#folder_name  = folder_name + [ "2019_" + "{:02d}".format(i) + "_ordersites" for i  in range(1,2) ]

data = {}

for dir_name in folder_name:
    entries = os.listdir(dir_name)
    for site_news in entries:
        working_directory = dir_name + "/" + site_news
        site = site_news[:-5]
    
        f = open(working_directory, "r")
        json_object = json.loads(f.read())
        f.close()
        for obj in json_object:
            amazon_accuracy, apple_accuracy = news_article_accuracy(obj)
            if site not in data:
                data[site] = {}
                data[site]["amazon_total"] = 0
                data[site]["amazon_accurate"] = 0
                
                data[site]["apple_total"] = 0
                data[site]["apple_accurate"] = 0
                
            if amazon_accuracy == 1 or amazon_accuracy == -1:
                data[site]["amazon_total"] += 1
            if amazon_accuracy == 1:
                data[site]["amazon_accurate"] += 1
            if apple_accuracy == 1 or apple_accuracy == -1:
                data[site]["apple_total"] += 1
            if apple_accuracy == 1:
                data[site]["apple_accurate"] += 1
            
```
